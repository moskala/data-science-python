```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
collection_name = ["gaming", "boardgames", "chess", "poker"]
df_gaming = pd.read_csv(os.path.join("results", "locations_{}.csv".format(collection_name[0])))
df_boardgames = pd.read_csv(os.path.join("results", "locations_{}.csv".format(collection_name[1])))
df_chess = pd.read_csv(os.path.join("results", "locations_{}.csv".format(collection_name[2])))
df_poker = pd.read_csv(os.path.join("results", "locations_{}.csv".format(collection_name[3])))
```

```python
df_gaming["GamingScore"] = list(range(20, 0, -1))
df_boardgames["BoardGamesScore"] = list(range(20, 0, -1))
df_chess["ChessGamesScore"] = list(range(20, 0, -1))
df_poker["PokerGamesScore"] = list(range(20, 0, -1))
```

```python
df_gaming = df_gaming[["CountryName", "GamingScore", "TotalUsers"]]
df_boardgames = df_boardgames[["CountryName", "BoardGamesScore", "TotalUsers"]]
df_chess = df_chess[["CountryName", "ChessGamesScore", "TotalUsers"]]
df_poker = df_poker[["CountryName", "PokerGamesScore", "TotalUsers"]]
```

```python
# df_gaming.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
join1 = df_gaming.merge(df_boardgames, how='outer', on="CountryName", suffixes=('_gaming', '_boardgames'))
join2 = df_chess.merge(df_poker, how='outer', on="CountryName", suffixes=('_chess', '_poker'))
```

```python
df = join1.merge(join2, how='outer', on="CountryName")
```

```python
df = df[["CountryName", "GamingScore", "BoardGamesScore", "ChessGamesScore", "PokerGamesScore", 
        "TotalUsers_gaming", "TotalUsers_boardgames", "TotalUsers_chess", "TotalUsers_poker"]]
```

```python
df
```

```python
df_all = df.dropna()
```

```python
df_all["TotalUsersAll"] = df_all["TotalUsers_gaming"] + df_all["TotalUsers_boardgames"] + df_all["TotalUsers_chess"] + df_all["TotalUsers_poker"]
```

```python
df_all
```

```python
df_all.to_csv("users_country_summary.csv")
```

```python
df_bar = df_all[["CountryName", "GamingScore", "BoardGamesScore", "ChessGamesScore", "PokerGamesScore"]]
df_bar.set_index(["CountryName"],inplace=True)
ax = df_bar.plot.bar()
ax.set_xlabel("Country Name")
ax.set_ylabel("Country Scores")
ax.set_title("Total users by country and topic")
```

```python
from pyecharts.charts import Pie, Grid
import pyecharts.options as opts

labels = df_all.CountryName
values = df_all.TotalUsersAll

pie = Pie()

pie.set_global_opts(title_opts=opts.TitleOpts(title='Total Users',
                                                    subtitle='Top 20 countries',
                                                    pos_left='right',
                                                    padding=(30, 1, 1, 1),
                                                    item_gap=10,
                                                    title_textstyle_opts=opts.TextStyleOpts(color='darkblue',
                                                                                            font_weight='bold',
                                                                                            font_family='Courier New',
                                                                                            font_size=30),
                                                    subtitle_textstyle_opts=opts.TextStyleOpts(color='grey',
                                                                                               font_weight='bold',
                                                                                               font_family='Courier New',
                                                                                               font_size=13)),
                          legend_opts=opts.LegendOpts(is_show=True, item_gap=10, padding=(0,0,50,0)))
pie.add("UsersTotal", list(zip(labels, values)))
pie.render_notebook() 
```

```python
pie.render("results\\pie_users_summary.html")
```

```python
df_sorted = df_bar.reset_index()
df_sorted["SumScore"] = df_sorted.GamingScore + df_sorted.BoardGamesScore + df_sorted.ChessGamesScore + df_sorted.PokerGamesScore
```

```python
df_sorted = df_sorted.sort_values(by=["SumScore"], ascending=False).reset_index(drop=True)
df_sorted
```

```python
df_sorted.to_csv("users_all_country.csv")
```
